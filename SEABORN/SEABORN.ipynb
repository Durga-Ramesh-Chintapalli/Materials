# Seaborn

Seaborn is an amazing data visualization library for statistical graphics plotting in python. it provides beautiful default styles and colour paletters to make statistical plots more attractive. It is built on the top of the matplotlib library and aslo closely integrated to the data structures from pandas.

# Import datasets

Seaborn has few important datasets in the library. When seaborn installed. The datasets download automatically, we can download any datasets with ""load_dataset()"" 

in this section, we will import a dataset by default

```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
```

```python
print(tips)
```

```python
tips.head()
```

# Plotting Univatiant Distribution with displot()

The most importent thing to do while analysing the data is to inderstand its distribution. Seaborn helps us to understand the univariant distribution of data.
The ""distplot()"" provides a quick look at univariant distribution. This function will plot a histogram that fits the kernel density estimate of the data.


```python
sns.distplot(tips['total_bill'])
plt.show()
```


# Seaborn Histogram

Histogram represent the data disribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin.
we can use same distplot() to plot a histogram as follows

```python
sns.displot(tips['total_bill'],kde=False)
```

# Seaborn- Kernal Density Estimates

Kernal Density Estimates(KDE) is a way to estimate the probabilty density function of a continuous random variable.it is used for non-parametric analysis.
setting the hist parameter to false in displot()function will yeild the kernel density estimate plot.

```python
sns.distplot(tips['total_bill'],hist=False)
plt.show()
```

# Plotting Bivariant Distibution with jointplot()

Bivariant Distibution is used to determain the relation b/w two variables. This mainly deals with relatonship b/w 2 variables and how one variable is behaving with respect to the other.
The best way to analyze Bivariant Distribution in seaborn is by using the jointplot() function.
jointplot() creates a multi panel figure that project the bivariant relationship b/w 2 variables and also the univariant distribution of each variable on separate axex.

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
plt.show()
```

# Seaborn-Scatter plot

a scatter plot can be used to demonstrate relationship b/w 2 variables x and y. A simple scatter plot can be drawn as follow.

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.show()
```

# Visualizing pairwise relationship with pairplot()


Some datasets contain many variables. In such cases, the relationship between each and every variable should be analyzed. So, we need to plot pairwise relationships in a dataset.

To plot multiple pairwise bivariate distributions, we can use pairplot() function. This shows the relationship for (n.2) combination of variable in a dataframe as a matrix of plots and the diagonal plots are the univariate plots

We can plot a pairplot as as follows:-


```python
sns.set_style("ticks")
sns.pairplot(tips)
```

# 17. Plotting categorical data

So far, I have covered histogram, scatter plot and kde plots. They are used to analyze the continuous variables under study. These plots are not suitable when the variable under study is categorical.

When one or both the variables under study are categorical, we can use plots like striplot() and swarmplot() to plot categorical data.

# 18. Seaborn Strip Plot
A stripplot() is used to draw a scatterplot where one variable is categorical. It represents the data in sorted order along any one of the axis.

We can plot a stripplot as follows:-


```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=False)
plt.show()
```

# Seaborn-Swam Plot

Another option which can be used as an alternate to jitter is to use function swarmplot(). This function positions each point of scatter plot the categorical axis and thereby avoids overlapping points.

So swarmplot() can be used to draw categorical scatterplot with non-overlapping points.

We can plot a swarmplot as follows:-

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
plt.show()
```
