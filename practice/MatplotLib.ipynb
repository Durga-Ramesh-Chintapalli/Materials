#  MatplotLib 

# Matplot lib is a plotting library for the python programming language 
# and its numerical mathematics extension Numpy.
It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter,WXPython, QT, or GTK+.

Some of the major pros of Matplotlib are:

***. Generally easy to get started for simple plots
. Support for Custom labels and texts
. Great control pf every element in a figure
. High - quality ouput in many formats
. very customizable in general ***

```python
# compute the x and Y coordinates for points on a sine curve
x= np.arange(0,4 * np.pi,0.1)
y = np.sin(x)
plt.title("since wave form")

# plot the points using matplot lib

plt.plot(x,y)
plt.show()
```

```python
# subplot()
# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# set up a sub plot grid that has height 2 and width ,
# and set the first such subplot as active

plt.subplot(2,1,1)
# make the first plot
plt.plot(x,y_sin) # 'b--'
plt.title('sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2,1,2)
plt.plot(x,y_cos) # 'g--
plt.title('Cosine')

# show the fig
plt.show()
```

```python
# Bar plot

x=[2,8,10]
y=[11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='r')
plt.title('Bar Graph')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

# Histogram

# Box Plot using Matplotlib
A Box plot is to created to display the set of data values
having properties like
minimum,first quartile,median,third quartile and maximum.

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

# rectangle box plot

plt.boxplot(data,vert=False,patch_artist=True);
```


# Pie  Chart

```python
# Data to Plot

labels = 'python','C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors =['gold', 'YellowGreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0) 

# plot 
plt.pie(sizes, explode= explode, labels=labels, colors=colors,
autopct='%1.1f%%',shadow=True)

plt.axis('equal')
plt.show()
```
