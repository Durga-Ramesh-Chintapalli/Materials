# Correlation with Heatmap

A correlation heatmap uses colored cells,
to show a 2D correlation matrix(table) between 
two discrete dimensions or event types. 
it is very important in Feature Selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Join Plot

A join plot allows to study the relationship 
between 2 numerica variables.
The centeral chart display their correction,
it is usually a scatterplot, ahexbin plot, 
a 20 histogram or a 20 density plot.


# Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# Pair Plot
A 'Pairs plot' is also known as a scatterplot,
in which one variable in the same data row is matched with another variables
value, like this: Pairs points are just elaborations on this,
showing at variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

# Dist plot
Dist plot helps us to check the distributions of the columns features

```python
sns.distplot(df['tip'])
```

```python
sns.displot(df['tip'],kde=False,bins=10)
```

# Categorical Plots

Seaborn also head us in doing the analysis on Categorical Data points.

* boxplot
* viollnplot
* countplot
* bar plot

```python
## count plot
sns.countplot('day',data=df) 
```

```python
sns.countplot(y='sex',data=df)
```

```python
## Bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
df.head()
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='total_bill',y='day', hue='smoker', data=df)
```

# Viollin Plot

Violin plot helps us to see the both the distribution of
data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day', data=df,palette='rainbow')
```
